# Matplotlib labels and titles
With Pyplot, we can use the xlabel() and ylabel() functions to set a label for the x- and y-axis

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x,y)

plt.xlabel("Average pulse")
plt.ylabel("Calorie burnage")
```

```python
#creating title for the graph
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x,y)

plt.xlabel("Average pulse")
plt.ylabel("Calorie burnage")
plt.title("Sports watch data")
```

```python
#You can use the fontdict parameter in xlabel(), ylabel(), and title() to set font properties for the title and labels
font1={'family':'serif','color':'blue','size':20}
font2={'family':'serif','color':'darkred','size':10}
plt.title("Sports watch data",fontdict=font1,loc='left') #loc to position the title
plt.xlabel("Average pulse",fontdict=font2)
plt.ylabel("Calories",fontdict=font1)
plt.plot(x,y)
plt.show()
```

```python
#grid() function to add grids
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid(axis='x')#axis parameter to specify the axis of grid to be showed in the screen

plt.show()
```

```python
#using line properties
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)

plt.show()
```

# Display Multiple Plots
With the subplot() function you can draw multiple plots in one figure:

```python
#plot:1
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])

plt.subplot(1,2,1)
plt.plot(x,y)

#plot:2
x=np.array([1,2,5,10])
y=np.array([3,7,9,6])

plt.subplot(1,2,2)
plt.plot(x,y)

plt.show()
```

The subplot() Function
The subplot() function takes three arguments that describes the layout of the figure.

The layout is organized in rows and columns, which are represented by the first and second argument.

The third argument represents the index of the current plot.

```python
x=np.array([0,1,2,3])
y=np.array([7,8,9,10])

plt.subplot(2,1,1) #basically the third argument is the plot no like this is the first plot so 1
plt.plot(x,y)
plt.title('SALES') #similarly you can add title also the plots

x=np.array([4,5,6,7])
y=np.array([11,12,13,14])

plt.subplot(2,1,2) #just like that second plot is 2
plt.plot(x,y)
plt.title('FORCE')

plt.show()
```

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("SALES")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("INCOME")

plt.suptitle("MY SHOP")
plt.show()
```
